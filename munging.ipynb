## Make necessary imports

```python
import pandas as pd
import numpy as np
```

## Use Pandas read csv file

```python
report = pd.read_csv("report.csv")
```

## Clean data

The inital file size is 399 rows

In the report, instead of using NaN, when information is missing it is marked with an 'X'. To change this for later filtering, we must change the 'X' values to 'NaN'

After changing the 'X' values to 'NaN', we drop any rows that have a NaN value for the Offense Number, because this is an important column and filling this value with a forward fill or backward fill is not practical.

The file size after removing these rows is now 368

```python
len(report)
```

```python
report = report.replace('X', np.nan)
report = report.dropna(subset=['Offense Code'])
```

```python
len(report)
```

## Easier Reading

To get our data down to the essentials, we'll drop comlumns that aren't necessary.

**Offense Code Extension**: Offense code is available

**Summary Offense Code**: Offense code is available

**Occurred Date or Date Range Start**: Date reported is available

**Occurred Date Range End**: Date reported is available

**District/Sector**: Hundred Block Location available

**Zone/Beat**: Hundred Block Location available

**Census Tract 2000**: Hundred Block Location available

**Longitude**: Location includes Long/Lat

**Latitude**: Location includes Long/Lat

```python
report = report.drop(['Offense Code Extension',
             'Summary Offense Code',
             'Occurred Date or Date Range Start',
             'Occurred Date Range End',
             'District/Sector',
             'Zone/Beat',
             'Census Tract 2000',
             'Longitude',
             'Latitude'], axis=1)
```
