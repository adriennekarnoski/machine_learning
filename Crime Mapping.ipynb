```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
report = pd.read_csv("report.csv")
```

```python
report
```

```python
long = report.iloc[:,(14)]
lat = report.iloc[:,(15)]
offense = report.iloc[:,(6)]
```

```python
print(len(offense))
```

```python
pers = ['ASSAULT',
        'DISPUTE',
        'HOMICIDE',
        'ROBBERY',
        'THREATS']

prop = ['BURGLARY-SECURE PARKING-RES',
        'CAR PROWL',
        'LOST PROPERTY',
        'OTHER PROPERTY',
        'PROPERTY DAMAGE',
        'STOLEN PROPERTY',
        'VEHICLE THEFT',
        'BIKE THEFT',
        'BURGLARY',
        'MAIL THEFT',
        'PICKPOCKET',
        'PURSE SNATCH']


beh = ['COUNTERFEIT',
       'FRAUD',
       'TRESPASS',
       'VIOLATION OF COURT ORDER',
       'WARRANT ARREST',
       'THEFT OF SERVICES',
       'EMBEZZLE',
       'ELUDING',
       'SHOPLIFTING',
       'ANIMAL COMPLAINT',
       'DISTURBANCE',
       'NARCOTICS',
       'WEAPON']
```

```python
color_list = []
offense_list = []
for item in offense:
    if item in pers:
        color_list.append('blue')
        if 'Offense Against a Person' in offense_list:
            offense_list.append('_')
        else:
            offense_list.append('Offense Against a Person')
    if item in prop:
        color_list.append('red')
        if 'Offense Against Property' in offense_list:
            offense_list.append('_')
        else:
            offense_list.append('Offense Against Property')
    if item in beh:
        color_list.append('green')
        if 'Criminal Behvaior' in offense_list:
            offense_list.append('_')
        else:
            offense_list.append('Criminal Behvaior')
```

```python
fig, ax = plt.subplots()
for i in range(len(offense)):
    ax.scatter(long[i], lat[i], c=color_list[i], s=100, label=offense_list[i], alpha=0.5, edgecolors='none')

plt.xlabel("Longitude")
plt.ylabel("Latitude")
```

```python
plt.grid(True)
plt.rcParams['figure.figsize'] = (11, 4)
plt.legend(loc='best')
plt.show()
```

```python
def distance(p, q):
    return np.sqrt(np.sum((p - q)**2))
```
